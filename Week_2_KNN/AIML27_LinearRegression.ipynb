```python
#import all required libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("/content/auto-mpg.data",names=["mpg","cylinders","displacement","horsepower","weight","acceleration","model_year","origin","car_name"],sep="\s+",na_values='?')
```

```python
lr.fit(X_train,y_train)
```

```python
y_predict = lr.predict(X_test)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(X_train,y_train)
```

```python
import joblib
joblib.dump(lr,"mymodel")
joblib.dump(scaler,"scalermodel")
```

```python
import gradio as gr
import joblib
def predict_mpg(cylinders,displacement,horsepower,weight,acceleration):
  input_data = [[cylinders,displacement,horsepower,weight,acceleration]]
  model = joblib.load("/content/mymodel")
  scaler = joblib.load("/content/scalermodel")
  input_scaled = scaler.transform(input_data)
  prediction = model.predict(input_scaled)
  return round(prediction[0],2)
app = gr.Interface(
    fn=predict_mpg,inputs=["number","number","number","number","number"],outputs="number",title="Auto MPG Prediction"
)
app.launch()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df['variety'].unique()
```

```python
df.columns
```

```python
X = df[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
y = df['variety']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
```

```python
y_predict = knn.predict(X_test)
```

```python
knn.score(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
for i in range(3,30):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train,y_train)
  print(f"score for neighbors {i} is {knn.score(X_test,y_test)}")
```

```python
import joblib
joblib.dump(knn,"knnmodel")
joblib.dump(scaler,"scalermodel")
```
